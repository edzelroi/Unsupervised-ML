```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Data/force2020_data_unsupervised_learning.csv', index_col='DEPTH_MD')
```

TRANSFORM DATA

```python
scaler = StandardScaler()
```

```python
df[['RHOB_T', 'NPHI_T', 'GR_T', 'PEF_T', 'DTC_T']] = scaler.fit_transform(df[['RHOB', 'NPHI', 'GR', 'PEF', 'DTC']])
```

```python
df
```

```python
def optimise_k_means(data, max_k):
    means = []
    inertias = []
    
    for k in range(1, max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        
        means.append(k)
        inertias.append(kmeans.inertia_)
        
        fig = plt.subplots(figsize = (10, 5))
        plt.plot(means, inertias, 'o-')
        plt.xlabel('Number of Clusters')
        plt.ylabel('Inertia')
        plt.grid(True)
        plt.show()
```

```python
optimise_k_means(df[['RHOB_T', 'NPHI_T']], 10)
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
df['kmeans_3'] = kmeans.labels_
```

```python
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df[['RHOB_T', 'NPHI_T']])
    df[f'kmeans_{k}'] = kmeans.labels_
```

```python
df
```

```python
fig, axs = plt.subplots(1, 5, figsize=(15, 3))

for k, ax in enumerate(fig.axes , start=1):
    ax.scatter(x=df['NPHI'], y=df['RHOB'], c= df[f'kmeans_{k}'])
    ax.set_title(f'N Clusters {k}')
```
